```python
import pandas as pd
dataset = pd.read_csv("Social_Network_ads.csv")
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset = dataset.drop(['User ID'],axis=1)
dataset.columns
independent = dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
dependent = dataset[['Purchased']]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import GridSearchCV
grid_param = {
    'alpha': [1.0],
    'force_alpha': [True, False],
    'fit_prior': [True, False],
    'class_prior': [None],
    'min_categories': [None]
    
}
grid = GridSearchCV(CategoricalNB(), grid_param, refit = True, verbose = 3,n_jobs=-1,scoring='f1')
grid.fit(X_train, y_train)
```

```python
grid_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, grid_pred)
```

```python
from sklearn.metrics import classification_report
cf_report = classification_report(y_test, grid_pred)
```

```python
print(cf_matrix)
print(cf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, grid_pred, average = 'weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```
