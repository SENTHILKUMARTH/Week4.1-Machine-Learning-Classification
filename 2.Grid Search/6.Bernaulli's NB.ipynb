```python
import pandas as pd
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset = dataset.drop(['User ID'],axis=1)
dataset.columns

#independent, dependent variable
independent = dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
dependent = dataset[['Purchased']]

#split training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)

#preprocessing the training and test set
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import GridSearchCV
grid_param = {
    'alpha': [0.1, 0.5, 1.0, 2.0],
    'binarize': [0.0, 0.5, 1.0]
}
grid = GridSearchCV(BernoulliNB(), grid_param, refit=True, verbose=3,scoring='f1_weighted')
grid.fit(X_train, y_train)
```

```python
grid_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, grid_pred)
```

```python
from sklearn.metrics import classification_report
cf_report = classification_report(y_test, grid_pred)
```

```python
print(cf_matrix)
```

```python
print(cf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, grid.predict_proba(X_test)[:,1])
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_pred, average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```
