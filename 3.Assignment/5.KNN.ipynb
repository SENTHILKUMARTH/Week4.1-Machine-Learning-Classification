```python
import pandas as pd
dataset = pd.read_csv("CKD.csv")
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset.columns

#independent, dependent variable
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent = dataset[['classification_yes']]

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
independent = sc.fit_transform(independent)
independent = sc.transform(independent)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
grid_param={
    'weights':['uniform', 'distance'],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute']
}

grid = GridSearchCV(KNeighborsClassifier(),grid_param, refit=True, verbose=3,scoring='explained_variance')
grid.fit(independent, dependent)
```

```python
grid_pred = grid.predict(independent)
```

```python
from sklearn.metrics import confusion_matrix
cf_metrics = confusion_matrix(dependent, grid_pred)
```

```python
from sklearn.metrics import classification_report
cf_report = classification_report(dependent, grid_pred)
```

```python
print(cf_metrics)
print(cf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(dependent, grid.predict_proba(independent)[:,1])
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(dependent,grid_pred, average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```
